```python


html = requests.get("https://procurement-notices.undp.org/").text
soup = BeautifulSoup(html, "lxml")
gdp = soup.find_all("table", attrs={"class": "standard cellborder"})
print("Number of tables on site: ", len(gdp))
table1 = gdp[0]
body = table1.find_all("tr")
head = body[0]
body_rows = body[1:]

headings = ["logo"]
for item in head.find_all("th"):
    item = item.text.rstrip("\n")
    headings.append(item)

all_rows = []
for row_num in range(len(body_rows)):
    row = []
    for row_item in body_rows[row_num].find_all("td"):
        if row_item.find("a") is not None:
            link = row_item.find("a").get("href")
            row.append(link)
        else:
            aa = re.sub("(\xa0)|(\n)|,", "", row_item.text)
            row.append(aa)
    all_rows.append(row)

data = pd.DataFrame(data=all_rows, columns=headings)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


def parse_data(listing_items):
    listing_list = []
    for idx in range(len(listing_items)):
        indv_listing = []

        current_listing = listing_items[idx]

        href = current_listing.select('link')[0].text if current_listing.select('link') else ""
        addy = current_listing.select('title')[0].get_text().replace('d&#039', '').replace(';', '; ') if current_listing.select('title') else ""
        hood = current_listing.select('description')[0].get_text().replace('<p>', '').replace('</p>\n', '').replace('\xa0;</li>\n<li>', ''). \
            replace('<a', '').replace('<ul><li>', '').replace('</li>\n</ul>', '').replace('</p>', ''). \
            replace('</a>', '').replace('">', '" - ') if current_listing.select('description') else ""
        pubdate = current_listing.select('pubDate')[0].string.replace('+0000', '') if current_listing.select('pubDate') else ""
        guid = current_listing.select('guid')[0].text if current_listing.select('guid') else ""
        indv_listing.append(href)
        indv_listing.append(addy)
        indv_listing.append(hood)
        indv_listing.append(pubdate)
        indv_listing.append(guid)

        listing_list.append(indv_listing)

    return listing_list


# Envoyer une requête GET à l'URL
r = requests.get('https://www.afdb.org/en/projects-and-operations/procurement.xml')
soup = BeautifulSoup(r.content, 'xml')
listing_items = soup.select('item')

# Extraire les données des projets
all_pages_parsed = parse_data(listing_items)

# Récupérer les liens dynamiques
dynamic_links = []
for item in all_pages_parsed:
    link = item[0]
    response = requests.get(link)
    if response.status_code == 200:
        dynamic_link = response.url
        dynamic_links.append(dynamic_link)
    else:
        dynamic_links.append("N/A")

# Ajouter les liens dynamiques au dataframe
df = pd.DataFrame(all_pages_parsed, columns=['link', 'title', 'description', 'pubDate', 'guid'])
df['dynamic_link'] = dynamic_links

# Afficher les données extraites avec les liens dynamiques
print(df)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

urls = [
    "https://ausschreibungen.giz.de/Satellite/company/welcome.do?fbclid=IwAR3Li8cr7eks2MIjJouywIrr6aynV43iSScpLXGYPIdaBlCNehtqFg93qbQ",
    "https://ausschreibungen.giz.de/Satellite/company/welcome.do?method=showTable&fromSearch=1&tableSortPROJECT_RESULT=2&tableSortAttributePROJECT_RESULT=publicationDate&selectedTablePagePROJECT_RESULT=2",
    "https://ausschreibungen.giz.de/Satellite/company/welcome.do?method=showTable&fromSearch=1&tableSortPROJECT_RESULT=2&tableSortAttributePROJECT_RESULT=publicationDate&selectedTablePagePROJECT_RESULT=3",
    "https://ausschreibungen.giz.de/Satellite/company/welcome.do?method=showTable&fromSearch=1&tableSortPROJECT_RESULT=2&tableSortAttributePROJECT_RESULT=publicationDate&selectedTablePagePROJECT_RESULT=4",
    "https://ausschreibungen.giz.de/Satellite/company/welcome.do?method=showTable&fromSearch=1&tableSortPROJECT_RESULT=2&tableSortAttributePROJECT_RESULT=publicationDate&selectedTablePagePROJECT_RESULT=5",
    "https://ausschreibungen.giz.de/Satellite/company/welcome.do?method=showTable&fromSearch=1&tableSortPROJECT_RESULT=2&tableSortAttributePROJECT_RESULT=publicationDate&selectedTablePagePROJECT_RESULT=6",
    "https://ausschreibungen.giz.de/Satellite/company/welcome.do?method=showTable&fromSearch=1&tableSortPROJECT_RESULT=2&tableSortAttributePROJECT_RESULT=publicationDate&selectedTablePagePROJECT_RESULT=7"
]

# List to store the data from all URLs
all_data = []

# Iterate over each URL
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the table element
    table = soup.find("table")

    # Extract column names from the table header
    header_row = table.find("thead").find("tr")
    column_names = [th.text.strip() for th in header_row.find_all("th")]

    # Extract data rows from the table body
    data_rows = table.find("tbody").find_all("tr")

    # Iterate over data rows and extract values for each column
    for row in data_rows:
        cells = row.find_all("td")
        row_data = [cell.text.strip() for cell in cells[:-1]]  # Exclude the last cell for "Aktion" column
        action_cell = cells[-1]  # Last cell for "Aktion" column
        link = action_cell.find("a")
        if link:
            row_data.append(link["href"])  # Extract the URL from the link
        else:
            row_data.append("")  # If no link, append an empty string
        all_data.append(row_data)

# Create a DataFrame from the combined data with column names
df = pd.DataFrame(all_data, columns=column_names)

# Print the DataFrame
print(df)
```
